```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('city_temperature.csv')
```

```python
df.head()
```

```python
def clean_and_calculate_avg_temperature(df):
    # make sure year column is integer
    df['Year'] = df['Year'].astype(int)
    
    # Replace unrealistic placeholder values (e.g., -99.0) with NaN
    df['AvgTemperature'] = df['AvgTemperature'].replace(-99.0, pd.NA)
    
    # Drop rows with NaN values in the AvgTemperature column
    df = df.dropna(subset=['AvgTemperature'])
    
    # Filter out unrealistic temperature values (e.g., below -80 or above 120)
    df = df[(df['AvgTemperature'] > -80) & (df['AvgTemperature'] < 120)]
    
    # Group by country and year, then calculate the average temperature
    country_yearly_avg_temp = df.groupby(['Country', 'Year'])['AvgTemperature'].mean().reset_index()
    
    # Rename the temperature column to something meaningful
    country_yearly_avg_temp.rename(columns={'AvgTemperature': 'YearlyAvgTemperature'}, inplace=True)
    
    return country_yearly_avg_temp
```

```python
country_yearly_avg_temp = clean_and_calculate_avg_temperature(df)

# Display the result
country_yearly_avg_temp 
```

```python
country_yearly_avg_temp.to_csv('country_yearly_avg_temp.csv', index=False)
```
